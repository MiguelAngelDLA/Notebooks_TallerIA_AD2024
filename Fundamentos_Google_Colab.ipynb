## Visualizaciones en Colab


Colab es compatible con bibliotecas populares de Python para la visualización de datos, como Matplotlib, Seaborn y Plotly.

Aquí tienes un ejemplo usando Matplotlib:


```python
# Ejemplo: Gráfico Simple con Matplotlib
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title("Onda Sinusoidal")
plt.xlabel("X")
plt.ylabel("sin(X)")
plt.show()
```
