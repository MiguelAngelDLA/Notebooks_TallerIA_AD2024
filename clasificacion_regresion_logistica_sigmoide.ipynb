# Clasificación con Regresión Logística

La **regresión logística** es una técnica de clasificación que se utiliza para predecir la probabilidad de que una observación pertenezca a una de dos categorías posibles. A diferencia de la regresión lineal, la regresión logística está diseñada para trabajar con variables dependientes binarias o categóricas. En esta técnica, la función sigmoide se utiliza para mapear los valores de entrada a una probabilidad entre 0 y 1.

## Función Sigmoide

La función sigmoide es una función matemática que convierte cualquier valor real en un valor en el rango de 0 a 1. Es ampliamente utilizada en modelos de clasificación binaria, incluyendo la regresión logística, ya que permite interpretar la salida como una probabilidad.

La fórmula de la función sigmoide es:

# $ \sigma(x) = \frac{1}{1 + e^{-x}} $

Donde \( x \) es el valor de entrada, y $ \sigma(x) $ es el resultado transformado entre 0 y 1. En el contexto de la regresión logística, este valor representa la probabilidad de que una observación pertenezca a la clase positiva.

#Ejemplo: detección de tumores

Para ilustrar el uso de la regresión logística en un ejemplo práctico, podemos considerar un caso común en biomedicina: la detección de tumores. En este ejemplo, se entrena un modelo de regresión logística para predecir si un tumor es maligno o benigno basándose en ciertos atributos del tumor, como el tamaño, la textura, y la forma de las células.

Supongamos que contamos con un conjunto de datos donde cada fila representa una observación *tumor* y cada columna representa una característica *o atributo* del tumor. La variable objetivo *la clase* tiene dos posibles valores:

0 para tumores benignos
1 para tumores malignos
El objetivo del modelo es predecir esta clase con base en las características de cada tumor.

Pasos para Entrenar el Modelo
Cargar el conjunto de datos: Utilizaremos datos ficticios o un conjunto de datos común como el Breast Cancer Wisconsin dataset de scikit-learn.

Preprocesamiento de datos: Normalizar o escalar las características si es necesario para mejorar la convergencia del modelo.

Entrenar el modelo: Aplicar regresión logística, ajustando los pesos de los parámetros para maximizar la probabilidad de clasificación correcta.

Evaluación del modelo: Utilizar métricas como la exactitud *accuracy*, la precisión, la sensibilidad *recall* y el AUC-ROC para medir el desempeño del modelo en clasificar correctamente los tumores.



```python
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

data = load_breast_cancer()
X = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

modelo = LogisticRegression(max_iter=10000)
modelo.fit(X_train, y_train)

y_pred = modelo.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Exactitud del modelo: {accuracy:.2f}")

print("\nMatriz de confusión:")
print(confusion_matrix(y_test, y_pred))

print("\nInforme de clasificación:")
print(classification_report(y_test, y_pred))
```
